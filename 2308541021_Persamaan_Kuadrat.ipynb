# ***Menentukan persamaan kuadrat jika diketahui akar-akarnya***

```python
import cmath

a_21 = float(input("Masukkan nilai a: "))
b_21 = float(input("Masukkan nilai b: "))
c_21 = float(input("Masukkan nilai c: "))
if a_21!=0:
    print('Nilai yang anda masukkan dapat membentuk persamaan kuadrat')
else:
    print('Nilai a tidak boleh 0')
#menghitung diskriminan
D_21 = (b_21**2) - (4*a_21*c_21)
print("Nilai diskriminan adalah ", D_21)
#mencari solusi akar
akar_1_21 = (-b_21-cmath.sqrt(D_21))/(2*a_21)
akar_2_21 = (-b_21+cmath.sqrt(D_21))/(2*a_21)
#cetak penyelsaian
print("Persamaan Kuadratnya adalah ({}x^2 + {}x + {})".format(a_21,b_21,c_21))
print("Hasil dari persamaan adalah {0} dan {1}".format(akar_1_21,akar_2_21))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cmath

def kuadrat_akar(a_21, b_21, c_21):
#Mencari akar-akar persamaan kuadrat
    akar1_21 = (-b_21 + (b_21**2 - 4*a_21*c_21)**0.5) / (2*a_21)
    akar2_21 = (-b_21 - (b_21**2 - 4*a_21*c_21)**0.5) / (2*a_21)
    return akar1_21, akar2_21

def kuadrat_persamaan(akar1_21, akar2_21):
#Mencari persamaan kuadrat jika diketahui akar-akarnya
    a_21 = 10
    b_21 = -(akar1_21 + akar2_21)
    c_21 = akar1_21 * akar2_21
    return f"x^2 + {b_21}x + {c_21}"

def kuadrat_grafik(a_21, b_21, c_21):
#Mencari grafik fungsi kuadrat
    x_21 = np.linspace(-10, 10, 400)
    y_21 = a_21*x_21**2 + b_21*x_21 + c_21
    plt.plot(x_21, y_21)
    plt.xlabel('x_21')
    plt.ylabel('y_21')
    plt.title('Grafik Fungsi Kuadrat')
    plt.grid(True)
    plt.show()

def kuadrat_titik_puncak(a_21, b_21):
#Mencari titik puncak dan sumbu simetri
    x_puncak_21 = -b_21 / (2*a_21)
    y_puncak_21 = a_21*x_puncak_21**2 + b_21*x_puncak_21
    return x_puncak_21, y_puncak_21

def kuadrat_posisi_puncak(a_21, b_21):
#Mencari posisi titik puncak
    if b_21 > 0:
        return "Titik puncak berada di kiri sumbu y"
    elif b_21 < 0:
        return "Titik puncak berada di kanan sumbu y"
    else:
        return "Titik puncak berada pada sumbu y"

def kuadrat_diskriminan(a_21, b_21, c_21):
#Mencari diskriminan persamaan kuadrat
    D_21 = b_21**2 - 4*a_21*c_21
    if D_21 > 0:
        return "Akar-akar persamaan kuadrat adalah nyata dan berbeda"
    elif D_21 == 0:
        return "Akar-akar persamaan kuadrat adalah nyata dan sama"
    else:
        return "Akar-akar persamaan kuadrat adalah kompleks"

def kuadrat_rumus_abc(a_21, b_21, c_21):
#Mencari akar-akar persamaan kuadrat menggunakan rumus ABC
    akar1_21 = (-b_21 + (b_21**2 - 4*a_21*c_21)**0.5) / (2*a_21)
    akar2_21 = (-b_21 - (b_21**2 - 4*a_21*c_21)**0.5) / (2*a_21)
    return akar1_21, akar2_21

def kuadrat_baru(akar1_21, akar2_21):
#Mencari persamaan kuadrat baru jika diketahui akar-akarnya
    a_baru_21 = 5
    b_baru_21 = -(akar1_21 + akar2_21)
    c_baru_21 = akar1_21 * akar2_21
    return f"x^2 + {b_baru_21}x + {c_baru_21}"


akar1_21, akar2_21 = kuadrat_akar(1, -5, 6)
print(f"Akar-akar persamaan kuadrat adalah x1 = {akar1_21} dan x2 = {akar2_21}")
persamaan_21 = kuadrat_persamaan(akar1_21, akar2_21)
print(f"Persamaan kuadrat adalah {persamaan_21}")
kuadrat_grafik(1, -5, 6)
x_puncak_21, y_puncak_21 = kuadrat_titik_puncak(1, -5)
print(f"Titik puncak adalah ({x_puncak_21}, {y_puncak_21})")
print(kuadrat_posisi_puncak(1, -5))
print(kuadrat_diskriminan(1, -5, 6))
akar1_21, akar2_21 = kuadrat_rumus_abc(1, -5, 6)
print("Akar-akar persamaan kuadrat menggunakan rumus ABC")
```
