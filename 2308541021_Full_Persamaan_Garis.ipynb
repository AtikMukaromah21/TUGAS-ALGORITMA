```python
import numpy as np
import matplotlib.pyplot as plt

kasus_21 = float(input("Masukkan kasus yang akan anda kerjakan (1/2/3) : "))

if kasus_21==1:
    print("=====Garis yang melalui dua titik.=====")
    #garis pertama
    a_21 = float(input("Nilai x dari titik pertama adalah "))
    b_21 = float(input("Nilai y dari titik pertama adalah "))
    c_21 = float(input("Nilai x dari titik kedua adalah "))
    d_21 = float(input("Nilai y dari titik kedua adalah "))
    #garis kedua
    e_21 = float(input("Nilai x dari titik pertama adalah "))
    f_21 = float(input("Nilai y dari titik pertama adalah "))
    g_21 = float(input("Nilai x dari titik kedua adalah "))
    h_21 = float(input("Nilai y dari titik kedua adalah "))

#gradien garis pertama    
m_1_21 = (d_21 - b_21)/(c_21 - a_21)
#gradien garis kedua
m_2_21 = (h_21 - f_21)/(g_21 - e_21)

#persamaan garis pertama  y_21 = m_1_21(x - a_21) + b_21 = m_1_21x - m_1_21a_21 + b_21
#persamaan garis kedua  y_21 = m_2_21(x - e_21) + f_21 = m_2_21x - m_2_21e_21 + f_21
#titik potong garis pertama terhadap sb-y
c_1_21 = -m_1_21*a_21+b_21
#titik potong garis kedua terhadap sb-y
c_2_21 = -m_2_21*e_21+f_21

print("Persamaan garis pertama adalah y={:.3f}x+{:.3f}".format(m_1_21,c_1_21))
print("Persamaan garis kedua adalah y={:.3f}x+{:.3f}".format(m_2_21,c_2_21))

if m_1_21==m_2_21:
    if c_1_21 == c_2_21:
        print("Kedua garis berhimpitan, sehingga tidak mempunyai solusi atau solusinya banyak.")
        x_21 = np.linspace(-10,10,100)
        plt.plot(x_21, m_1_21*(x_21 - a_21)+b_21,color='blue')
        plt.plot(x_21, m_2_21*(x_21 - e_21)+f_21,color='purple')
        plt.axvline(x_21 = 0,color='red')
        plt.axhline(y_21 = 0,color='yellow')
        plt.show()
    else:
        print("Kedua garis tidak mungkin berpotongan, sehingga tidak mempunyai solusi.")
        x_21 = np.linspace(-10,10,100)
        plt.plot(x_21, m_1_21*(x_21 - a_21)+b_21,color='red')
        plt.plot(x_21, m_2_21*(x_21 - e_21)+f_21,color='green')
        plt.axvline(x_21 = 0,color='black')
        plt.axhline(y_21 = 0,color='grey')
        plt.show()
else:
    print("Terdapat satu titik potong")
    tp_x_21 = (-m_2_21 * e_21 +f_21 + m_1_21 * a_21 - b_21)/(m_1_21 - m_2_21)
    tp_y_21 = m_1_21 * (tp_x_21 - a_21) + b_21
    print("Maka titik potong kedua garis adalah ({:.3f},{:.3f})".format(tp_x_21,tp_y_21))

    x_21 = np.linspace(tp_x_21-10,tp_x_21 + 10,100)
    plt.plot(x_21,m_1_21*(x_21-a_21)+b_21,color='pink',label="y={:.3f}(x-{:.3f})+{:.3f}".format(m_1_21,a_21,b_21))
    plt.plot(x_21,m_2_21*(x_21-e_21)+f_21,color='brown',label="y={:.3f}(x-{:.3f})+{:.3f}".format(m_2_21,e_21,f_21))
    plt.axvline(x=0,color='black')
    plt.axhline(y=0,color='black')
    plt.legend(loc='best')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

kasus_21 = float(input("Masukkan kasus yang akan anda kerjakan (1/2/3) :"))

if kasus_21 == 1:
    print("=====Garis yang melalui dua titik.=====")
    #garis pertama
    a_21 = float(input("Nilai x dari titik pertama adalah "))
    b_21 = float(input("Nilai y dari titik pertama adalah "))
    c_21 = float(input("Nilai x dari titik kedua adalah "))
    d_21 = float(input("Nilai y dari titik kedua adalah "))
    #garis kedua 
    e_21 = float(input("Nilai x dari titik pertama adalah "))
    f_21 = float(input("Nilai y dari titik pertama adalah "))
    g_21 = float(input("Nilai x dari titik kedua adalah "))
    h_21 = float(input("Nilai y dari titik kedua adalah "))
    
    #gradien garis pertama
    m_1_21 = (d_21 - b_21) / (c_21 - a_21)
    #gradien garis kedua
    m_2_21 = (h_21 - f_21) / (g_21 - e_21)
    #titik potong garis pertama terhadap sb-y
    c_1_21 = -m_1_21 * a_21 + b_21
    #titik potong garis kedua terhadap sb-y
    c_2_21 = -m_2_21 * e_21 + f_21
    
    print("Persamaan garis pertama adalah y={:.3f}x+{:.3f}".format(m_1_21, c_1_21))
    print("Persamaan garis kedua adalah y={:.3f}x+{:.3f}".format(m_2_21, c_2_21))

    if m_1_21 == m_2_21:
        if c_1_21 == c_2_21:
            print("Kedua garis berhimpitan, sehingga tidak mempunyai solusi atau solusinya banyak.")
            x_21 = np.linspace(-5, 5, 100)
            plt.plot(x_21, m_1_21 * (x_21 - a_21) + b_21, color='blue')
            plt.plot(x_21, m_2_21 * (x_21 - e_21) + f_21, color='yellow')
            plt.axvline(x=0, color='black')
            plt.axhline(y=0, color='black')
            plt.show()
        else:
            print("Kedua garis tidak mungkin berpotongan, sehingga tidak mempunyai solusi atau solusinya banyak.")
            x_21 = np.linspace(-5, 5, 100)
            plt.plot(x_21, m_1_21 * (x_21 - a_21) + b_21, color='blue')
            plt.plot(x_21, m_2_21 * (x_21 - e_21) + f_21, color='yellow')
            plt.axvline(x=0, color='black')
            plt.axhline(y=0, color='black')
            plt.show()
    else:
        print("Terdapat satu titik potong")
        tp_x_21 = (-m_2_21 * e_21 + f_21 + m_1_21 * a_21 - b_21) / (m_1_21 - m_2_21)
        tp_y_21 = m_1_21 * (tp_x_21 - a_21) + b_21
        print("Maka titik potong kedua garis adalah ({:.3f},{:.3f})".format(tp_x_21, tp_y_21))
    
        x_21 = np.linspace(tp_x_21 - 5, tp_x_21 + 5, 100)
        plt.plot(x_21, m_1_21 * (x_21 - a_21) + b_21, color='blue', label="y={:.3f}(x-{:.3f})+{:.3f}".format(m_1_21, a_21, b_21))
        plt.plot(x_21, m_2_21 * (x_21 - e_21) + f_21, color='yellow', label="y={:.3f}(x-{:.3f})+{:.3f}".format(m_2_21, e_21, f_21))
        plt.scatter([tp_x_21], [tp_y_21], color='red', zorder=5)
        plt.text(tp_x_21, tp_y_21, '({:.2f}, {:.2f})'.format(tp_x_21, tp_y_21), fontsize=14, ha='right')
        plt.axvline(x=0, color='black')
        plt.axhline(y=0, color='black')
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
        
elif kasus_21== 2:
    print("=====Garis diketahui gradien dan yang melalui titik.=====")
    #garis pertama
    m_1_21 = float(input("Gradien dari garis pertama adalah "))
    a_21 = float(input("Nilai x dari titik yang dilewati adalah "))
    b_21 = float(input("Nilai y dari titik yang dilewati adalah "))
    print("Persamaan garis pertama adalah y={}*(x-{})+{}".format(m_1_21, a_21, b_21))
    
    #garis kedua
    m_2_21 = float(input("Gradien dari garis kedua adalah "))
    c_21 = float(input("Nilai x dari titik yang dilewati adalah "))
    d_21 = float(input("Nilai y dari titik yang dilewati adalah "))
    print("Persamaan garis kedua adalah y={}*(x-{})+{}".format(m_2_21, c_21, d_21))
    
    c_1_21 = -m_1_21 * a_21 + b_21
    c_2_21 = -m_2_21 * c_21 + d_21
    
    if m_1_21 == m_2_21:
        print("Kedua garis sejajar")
        if c_1_21 == c_2_21:
            print("Sehingga terdapat tak berhingga banyaknya solusi")
            x_21 = np.linspace(-5, 5, 100)
            plt.plot(x_21, m_1_21 * (x_21 - a_21) + b_21, color='blue', label='Garis pertama dan kedua (Berhimpitan)')
            plt.axvline(x=0, color='black', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='black', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("Tidak terdapat solusi")
            x_21 = np.linspace(-5, 5, 100)
            plt.plot(x_21, m_1_21 * (x_21 - a_21) + b_21, color='blue', label='Garis pertama')
            plt.plot(x_21, m_2_21 * (x_21 - c_21) + d_21, color='yellow', label='Garis kedua')
            plt.axvline(x=0, color='black', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='black', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("Kedua garis akan berpotongan di satu titik.")
        tp_x_21 = (c_2_21 - c_1_21) / (m_1_21 - m_2_21)
        tp_y_21 = m_1_21 * (tp_x_21 - a_21) + b_21
        print("Maka titik potong kedua garis adalah ({:.3f},{:.3f})".format(tp_x_21, tp_y_21))
        
        x_21 = np.linspace(tp_x_21 - 5, tp_x_21 + 5, 100)
        plt.plot(x_21, m_1_21 * (x_21 - a_21) + b_21, color='blue', label="Garis pertama: y={:.3f}(x-{:.3f})+{:.3f}".format(m_1_21, a_21, b_21))
        plt.plot(x_21, m_2_21 * (x_21 - c_21) + d_21, color='yellow', label="Garis kedua: y={:.3f}(x-{:.3f})+{:.3f}".format(m_2_21, c_21, d_21))
        plt.scatter([tp_x_21], [tp_y_21], color='green', zorder=5)
        plt.text(tp_x_21, tp_y_21, '({:.2f}, {:.2f})'.format(tp_x_21, tp_y_21), fontsize=14, ha='right')
        plt.axvline(x=0, color='black', linestyle='--', linewidth=0.5)
        plt.axhline(y=0, color='black', linestyle='--', linewidth=0.5)
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

elif kasus_21 == 3:
    print("=====Kedua garis yang memenuhi persamaan berbentuk ax+by=c dan dx+ey=f.=====")
    #garis pertama
    a_21 = float(input("Koefisien dari x adalah "))
    b_21 = float(input("Koefisien dari y adalah "))
    c_21 = float(input("Konstanta dari garis pertama adalah "))
    #garis kedua
    d_21 = float(input("Koefisien dari x adalah "))
    e_21 = float(input("Koefisien dari y adalah "))
    f_21 = float(input("Konstanta dari garis kedua adalah "))
    
    det_21 = a_21 * e_21 - b_21 * d_21
    
    if det_21 == 0:
        print("Kedua garis akan sejajar atau berhimpitan")
        if (a_21 * f_21 - c_21 * d_21) == 0 and (b_21 * f_21 - c_21 * e_21) == 0:
            print("Kedua garis berhimpitan")
            x_21 = np.linspace(-5, 5, 100)
            y_21 = (c_21 - a_21 * x_21) / b_21
            plt.plot(x_21, y_21, color='blue', label='Garis berhimpitan')
            plt.axvline(x=0, color='black', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='black', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("Kedua garis sejajar")
            x_21 = np.linspace(-5, 5, 100)
            y1_21 = (c_21 - a_21 * x_21) / b_21
            y2_21 = (f_21 - d_21 * x_21) / e_21
            plt.plot(x_21, y1_21, color='blue', label='Garis pertama')
            plt.plot(x_21, y2_21, color='yellow', label='Garis kedua')
            plt.axvline(x=0, color='black', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='black', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("Kedua garis akan berpotongan di satu titik.")
        tp_x_21 = (c_21 * e_21 - b_21 * f_21) / det_21
        tp_y_21 = (a_21 * f_21 - c_21 * d_21) / det_21
        print("Maka titik potong kedua garis adalah ({:.3f},{:.3f})".format(tp_x_21, tp_y_21))
        
        x_21 = np.linspace(tp_x_21 - 5, tp_x_21 + 5, 100)
        y1_21 = (c_21 - a_21 * x_21) / b_21
        y2_21 = (f_21 - d_21 * x_21) / e_21
        plt.plot(x_21, y1_21, color='blue', label='Garis pertama')
        plt.plot(x_21, y2_21, color='yellow', label='Garis kedua')
        plt.scatter([tp_x_21], [tp_y_21], color='green', zorder=5)
        plt.text(tp_x_21, tp_y_21, '({:.2f}, {:.2f})'.format(tp_x_21, tp_y_21), fontsize=14, ha='right')
        plt.axvline(x=0, color='black', linestyle='--', linewidth=0.5)
        plt.axhline(y=0, color='black', linestyle='--', linewidth=0.5)
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
else:
    print("Jadi terdapat 3 kasus. Masukkan kasus yang anda inginkan 1, 2 atau 3 saja.")
```
