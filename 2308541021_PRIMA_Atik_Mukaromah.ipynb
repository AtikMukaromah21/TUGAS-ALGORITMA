# **Mencari apakah n bilangan prima**

```python
import math

n_21 = int(input("Masukkan angka yang diinginkan = "))
if n_21 <= 1:
    print("ini bukan bilangan prima")
elif n_21 == 2:
    print("ini adalah bilangan prima")
elif n_21 % 2 == 0:
    print("ini bukan bilangan prima")
else:
    prima_21 = True
    for i_21 in range(3, int(math.sqrt(n_21)) + 1, 2):
        if n_21 % i_21 == 0:
            prima_21 = False
            break
    if prima_21:
        print("Ini adalah bilangan prima")
    else:
        print("Ini bukan bilangan prima")
```

# **Jika diketahui bahwa 𝑛 adalah 21, maka tentukan bilangan prima antara 1000𝑛 dan 1000(𝑛+1)**

```python
import math

print("Jika diketahui bahwa 𝑛 adalah 21, maka tentukan bilangan prima antara 1000𝑛 dan 1000(𝑛+1).")
print("Akan dicari bilangan prima antara 21000 sampai 22000")
print("========================================================================================")

def bilangan_prima(angka_21):
    if angka_21 < 2:
        return False
    for i_21 in range(2, int(angka_21 ** 0.5) + 1):
        if angka_21 % i_21 == 0:
            return False
    return True

#jika diketahui
nim_21 = 21
a_21 = 1000 * nim_21
b_21 = 1000 * (nim_21 + 1)

prima_21 = [angka_21 for angka_21 in range(a_21, b_21) if bilangan_prima(angka_21)]

print("Jadi Bilangan Prima antara", a_21, "sampai", b_21, "adalah:")
print(prima_21)
```
