```python
nilai_21 = input("Berapakah nilai rata-rata Anda? ")
nilai_21 = float(nilai_21)
if nilai_21 == 100:
    peringkat_21 = 'A+'
    print("\tGOOD!")
elif nilai_21 >= 90.0:
    peringkat_21 = 'A+'
elif nilai_21 >= 80.0:
    peringkat_21 = 'A'
elif nilai_21 >= 70.0:
    peringkat_21 = 'B+'
elif nilai_21 >= 60.0:
    peringkat_21 = 'B'
elif nilai_21 >= 50.0:
    peringkat_21 = 'C+'
elif nilai_21 < 40.0:
    peringkat_21 = 'D'
    
print("Peringkat anda adalah " + peringkat_21)  
```
