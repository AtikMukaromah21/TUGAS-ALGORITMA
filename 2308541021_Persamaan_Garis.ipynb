```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan garis pertama melewati titik (a,b) dan (c,d)
a=-1 #Titik (a,b)
b=0
c=2 #Titik (c,d)
d=-2
m1=(d-b)/(c-a)
print("Gradien garis pertama adalah {:.2f}".format(m1))
if -m1*a+b>0:
    print("Persamaan garis pertama adalah y={:.2f}x+{:.2f} ".format(m1,-m1*a+b))
else:
    print("Persamaan garis pertama adalah y={:.2f}x{:.2f} ".format(m1,-m1*a+b))
#y1 = m1*(x-a)+b

# Persamaan garis kedua melewati titik (e,f) dan (g,h)
e=5 #Titik (e,f)
f=0
g=-5  #Titik (g,h)
h=-20
m2=(h-f)/(g-e)
print("Gradien garis ke dua adalah {:.2f}".format(m2))
if -m2*e+f>0:
    print("Persamaan garis kedua adalah y={:.2f}x+{:.2f} ".format(m2,-m2*e+f))
else:
    print("Persamaan garis kedua adalah y={:.2f}x{:.2f} ".format(m2,-m2*e+f))
#y2=m2*(x-e)+f

if m1==m2:
    print("Gradiennya sama lho, jadi kemungkinannya ga ada solusi atau solusinya banyak.")
else:
    # Titik potong kedua garis pada (tp_x,tp_y)
    tp_x = (m1*a-b-m2*e+f)/(m1-m2)
    tp_y = m2*(tp_x-e)+f

    print("Titik potong kedua garis adalah ({:.2f},{:.2f})".format(tp_x,tp_y))

    x=np.arange(tp_x-10,tp_x+10,0.01)
    plt.figure(figsize=(4,4),dpi=500)
    if -m1*a+b>0:
        plt.plot(x,m1*(x-a)+b,color='red',label='y={:.2f}x+{:.2f}'.format(m1,-m1*a+b))
    else:
        plt.plot(x,m1*(x-a)+b,color='red',label='y={:.2f}x{:.2f}'.format(m1,-m1*a+b))
    if -m2*e+f>0:
        plt.plot(x,m2*(x-e)+f,color='blue',label='y={:.2f}x+{:.2f}'.format(m2,-m2*e+f))
    else:
        plt.plot(x,m2*(x-e)+f,color='blue',label='y={:.2f}x{:.2f}'.format(m2,-m2*e+f))
        
    plt.plot(tp_x,tp_y,color='green',marker='o',markersize=4)
    plt.legend(loc='best')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#persamaan garis pertama melewati titik (p_21,q_21) dan (r_21,s_21)
#titik (p_21,q_21)
p_21 = 5 
q_21 = 3
#titik (r_21,s_21)
r_21 = 1
s_21 = 2
m1_21 = (s_21-q_21)/(r_21-p_21) 
print ("Gradien garis pertamanya adalah ".format(m1_21))

#mencari persamaan garis pertama y1 = m1*(x - p_21) + q_21
print("Persamaan garis pertama adalah y={}x+{}".format(m1_21,-m1_21*p_21+q_21)) 


#persamaan garis kedua melewati titik (t_21,u_21) dan (v_21,w_21)
#titik (t_21,u_21)
t_21 = 4
u_21 = -1
#titik (v_21,w_21)
v_21 = 3
w_21 = -1
m2_21 = (w_21-u_21)/(v_21-t_21)
print("Gradien garis kedua adalah {}".format(m2_21))

#mencari persamaan garis kedua y2 = m2*(x - t_21) + u_21 
print("Persamaan garis kedua adalah y={}x+{}".format(m2_21,-m2_21*t_21+u_21))


#mencari titik potong kedua garis 
if m1_21==m2_21:
    print( "Gradien pertama dan kedua sama, sehingga tidak mempunyai solusi atau solusinya banyak.") 
else:
#mencari titik potong kedua garis (tp_x, tp_y)
    tp_x_21 = (m1_21*p_21-q_21-m2_21*t_21+u_21)/(m1_21-m2_21)
    tp_y_21 = m2_21*(tp_x_21-t_21)+u_21

    print ("Titik potong x dan y adalah ({},{})".format(tp_x_21,tp_y_21))

#mencari range dari x
    x_21=np.arange(tp_x_21-10,tp_x_21+5,0.01)
#mencari ukuran dan resolusi figure
    plt.figure(figsize=(4,4),dpi=120)
#mencari plot persamaan garis pertama
    plt.plot(x_21,m1_21*(x_21-p_21)+q_21,color='purple',label='y={}x+{}'.format(m1_21,-m1_21*p_21+q_21))
#mencari plot persamaan garis kedua
    plt.plot(x_21,m2_21*(x_21-t_21)+u_21,color='blue',label='y={}x+{}'.format(m2_21,-m2_21*t_21+u_21))
#mencari plot titik potong
    plt.plot(tp_x_21,tp_y_21,color='red',marker='o')
    plt.legend(loc="best") 
    plt.show()
    print("Perhatikan gambar berikut, persamaan garis pertama berwarna ungu, persamaan garis kedua berwarna biru dan titik potong berwarna merah")
```
